#  Hierarchical Clustering

```python
from sklearn import datasets
import pandas as pd
import numpy as np

import seaborn as sns
```

## IRIS DS

```python
iris = datasets.load_iris()
# iris.data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(iris.data,iris.target, test_size = 0.20, random_state = 166)
sns.scatterplot(x=X_train[:,0], y=X_train[:,-1], c =["green"])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
dendrogram(linkage(X_test, 'single'),orientation='top',distance_sort='descending',show_leaf_counts=True)

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
hc_model.fit_predict(X_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.scatter(x=X_test[:, 0],y=X_test[:, -1],c=['green'])
plt.scatter(x=X_test[:,0], y=X_test[:,1], c= hc_model.labels_, cmap='rainbow' )
plt.show()
```

## Diabetes ds

```python
diabetes = datasets.load_diabetes()
# diabetes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(diabetes.data,diabetes.target, test_size = 0.20, random_state = 166)
sns.scatterplot(x=X_train[:,0], y=X_train[:,-1], c =["green"])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram(linkage(X_test, 'single'),orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
hc_model.fit_predict(X_test)
```

```python
plt.scatter(x=X_test[:,0], y=X_test[:,1], c= hc_model.labels_, cmap='rainbow' )
```
