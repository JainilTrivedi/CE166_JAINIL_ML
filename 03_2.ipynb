# Apply algorithm on breast cancer wisconsin dataset - One Hot Encoding of features: and Train test Division 50%-50%

```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
# from sklearn import metrics
datasets = pd.read_csv("BreastCancerWiscoin.csv")
datasets = datasets.drop(['fractal_dimension_worst'], axis=1)
datasets
```

```python
onehotencoder = OneHotEncoder()
x = onehotencoder.fit_transform(datasets.diagnosis.values.reshape(-1,1)).toarray()
# print(x)
dfOneHot = pd.DataFrame(x, columns = ["diagnosis_"+str(int(i)) for i in range(2)]) 
df = pd.concat([datasets, dfOneHot], axis=1)
df= df.drop(['diagnosis'], axis=1) 
feature=df.iloc[:,:-2].values
out=df.iloc[:,-2:].values
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(feature, out, test_size = 0.50, random_state = 166)
# print(data_train.shape,target_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_train = sc.fit_transform(data_train)
data_test = sc.transform(data_test)
print(data_train)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier0 = GaussianNB()
# print(target_train[:,0])
# print(target_train[:,1])
classifier0.fit(data_train,target_train[:,0])
targetPred0 = classifier0.predict(data_test)
print(targetPred0)
```

```python
classifier1 = GaussianNB()
classifier1.fit(data_train,target_train[:,1])
targetPred1 = classifier1.predict(data_test)
print(targetPred1)
```

```python
target_pred = np.column_stack((targetPred0,targetPred1))
print(target_pred)
```

```python
from sklearn import metrics
print("Accuracy:\t",metrics.accuracy_score(target_test, target_pred))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(target_test, target_pred, average=None)
recall = recall_score(target_test, target_pred, average=None)
print('precision:\t{}'.format(precision))
print('recall:\t{}'.format(recall))
```
