## Corelation

```python
nds = ds 
# imputer = imputer.fit(nds[:,1:10])

nds = nds.drop(labels=['Unnamed: 0'],axis = 1)
nds
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corrmat = nds.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corrmat, annot=True, cmap="RdYlGn")
```

```python
columns = np.full((len(nds.columns),), True, dtype=bool)
for i in range(corrmat.shape[0]):
    for j in range(i+1, corrmat.shape[0]):
        if corrmat.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
selected_columns = nds.columns[columns]
# selected_columns.drop(['Unnamed: 0'],axis=1,inplace=True])
selected_columns
```
