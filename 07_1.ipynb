```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import math

warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv('BuyComputer.csv')
ds.drop(labels = ['User ID'],axis=1,inplace=True)
ds.head(7)
```

```python
X = ds.iloc[:,:-1].values
print(X)
Y = ds.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state = 166)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# X_test
# X_train
```

```python
y_pred = []
len_x = len(X_train[0])
w = []
b = 0.2
print(len_x)
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def predict(input):
    z = np.dot(input, w) + b
    h= sigmoid(z)
    for i in range(len(h)):
        if(h[i]>=0.5):
            h[i]=1
        else:
            h[i]=0
    return h

def loss_func(y, a):
    j=0
    total_loss = np.sum( (-a * np.log(y)) - ((1 - a) * np.log(1 - y)))
    m = y.shape[0]
    j = total_loss /m
    return j
```

```python
for i in range(3000):
    z = np.dot(X_train, w) + b
    y_pred = sigmoid(z)
    loss = loss_func(y_test, y_test)
    dw = np.dot((y_pred-y_train).T, X_train)/X_train.shape[0]
    db = np.mean(y_pred-y_train)
    w = w - alpha * dw
    b = b - alpha* db
    print(w,b,loss)
# len(y_train)
```

```python
y_pred = predict(X_test)
print(len(y_pred))
len(y_test)
```

```python
def accuracy(y_test,y_pred):
    c=0
    i=0
    for j in range(0,len(y_test)):
        if y_pred[j] == y_test[j]:
            c+=1
        else:
            i+=1
    return (float(c)/(len(y_test)))
```
