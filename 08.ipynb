# Bigram

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer2 = TfidfVectorizer(stop_words={'english'},ngram_range=(2,2))
```

```python
tfdif = vectorizer2.fit(ds['v2'])
print(tfdif)
tfdif = vectorizer2.transform(ds['v2'])
tfdif.toarray()
tfdif = vectorizer2.fit_transform(ds['v2'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(tfdif,ds['v1'],random_state=166)
dtc2 = DecisionTreeClassifier()
mnb2 = MultinomialNB()

dtc2.fit(X_train,y_train)
mnb2.fit(X_train,y_train)
tfdif_dt = dtc2.predict(X_test)
tfdif_mnb = mnb2.predict(X_test)

print("MNB")
from sklearn.metrics import precision_score,recall_score
print("Precision Score:", precision_score(y_test,tfdif_mnb))
print("Recall Score:", recall_score(y_test,tfdif_mnb))
print("DTC")
from sklearn.metrics import precision_score,recall_score
print("Precision Score:", precision_score(y_test,tfdif_dt))
print("Recall Score:", recall_score(y_test,tfdif_dt))
```
